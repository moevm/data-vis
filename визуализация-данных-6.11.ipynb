```python
import pandas_datareader as pdr
from pandas_datareader import wb
```

```python
countries = wb.get_countries()
df = wb.download(indicator="SP.DYN.LE00.IN", country=countries['iso2c'], start='1980', end='2020')
df.index.levels
```

```python
df.reset_index(inplace=True)
df.head()
```

```python
df.rename(columns={'SP.DYN.LE00.IN':'LSp'}, inplace=True)
df.head()
```

```python
df = df.pivot(index='country', columns='year', values='LSp')
df.head()
```

```python
df.isnull().sum().sum()
```

```python
df.dropna(how='all', axis=1, inplace=True)
df.dropna(how='all', inplace=True)
df.isnull().sum().sum()
```

```python
df.head()
```

```python
df['mean'] = df.mean(axis=1)
df.head()
```

```python
years = list(map(str, range(1980, 2019)))
Angola = df.loc['Angola', years]
Angola.plot()
```
