```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image
import folium
```

```python
text = ["red"] * 80 + ['blue']*50 + ["green"] * 10 + ['yellow']*2
text = ' '.join(text)
text
```

```python
wc = WordCloud(background_color="white")
wc.generate(text)
plt.axis("off")
plt.imshow(wc, interpolation="bilinear")
plt.show()
```

```python
words_dict = {'red': 80, 'blue':40, 'green':10, 'yellow':2}
wc = WordCloud(background_color="white")
wc.generate_from_frequencies(words_dict)
plt.axis("off")
plt.imshow(wc, interpolation="bilinear")
plt.show()
```

```python
fr_merc_mask = np.array(Image.open('freddie-mercury-5-png-1001573170861bk08infhnj.png'))
fr_merc_mask
```

```python
fig = plt.figure()
fig.set_figwidth(15) # set width
fig.set_figheight(15) # set height

plt.imshow(fr_merc_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
comp = np.array(fr_merc_mask[0, 0])

for i, row in enumerate(fr_merc_mask):
    for j, el in enumerate(row):
        if np.array_equal(el, comp):
            fr_merc_mask[i, j] = np.array([255, 255, 255, 255])
        else:
            fr_merc_mask[i, j] = np.array([0, 0, 0, 255])

fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(fr_merc_mask, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
freddie_merc_wc = WordCloud(background_color='black', max_words=2000, mask=fr_merc_mask, stopwords=STOPWORDS,contour_color='blue', contour_width=1, prefer_horizontal=0.3)
freddie_merc_wc.generate(fr_merc_songs)
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(freddie_merc_wc, cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
fig = plt.figure()
fig.set_figwidth(14) # set width
fig.set_figheight(18) # set height

plt.imshow(freddie_merc_wc.recolor(colormap='PuBu'), cmap=plt.cm.gray, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
df = pd.read_csv('meteorite-landings.csv')
df.head()
```

```python
df.rename(columns={'reclat':'lat', 'reclong':'long'}, inplace=True)
df.head()
```

```python
df.isnull().sum().sum()
```

```python
df.dropna(how='any', inplace=True)
```

```python
df = df.iloc[:100]
```

```python
lat = 50.775
lng = 6.08333

met_map = folium.Map(location=[lat, lng], zoom_start=4)

meteorite_group = folium.map.FeatureGroup()

for lat, lng in zip(df['lat'], df['long']):
    folium.CircleMarker(
        [lat, lng], 
        radius=5, 
        color='red', 
        fill=True, 
        fill_color='white', 
        popup='text',
    ).add_to(meteorite_group)

met_map.add_child(meteorite_group)
```

```python
met_map = folium.Map(location=[lat, lng], zoom_start=4)
labels = ['{} {}'.format(name, year)  for name, year in zip(list(df['name']), list(df['year']))]
for lat, lng, label in zip(df['lat'], df['long'], labels):
    folium.Marker(
        [lat, lng], 
        icon=folium.Icon(color='purple'),
        popup=label,
    ).add_to(met_map)

met_map
```

```python
from folium import plugins

met_map = folium.Map(location=[lat, lng], zoom_start=4)

met_cluster = plugins.MarkerCluster().add_to(met_map)


labels = ['{} {}'.format(name, year)  for name, year in zip(list(df['name']), list(df['year']))]

for lat, lng, label in zip(df['lat'], df['long'], labels):
    folium.Marker(
        [lat, lng], 
        icon=folium.Icon(color='purple'),
        popup=label,
    ).add_to(met_cluster)

met_map
```

```python
from pandas_datareader import wb
countries = wb.get_countries()
df = wb.download(indicator="SP.DYN.LE00.IN", country=countries['iso2c'], start='1980', end='2020')
df.reset_index(inplace=True)
df.rename(columns={'SP.DYN.LE00.IN':'lifespan'}, inplace=True)
df = df.pivot(index='country', columns='year', values='lifespan')
df.dropna(how='all', axis=1, inplace=True)
df.dropna(how='all', axis=0, inplace=True)
df.dropna(how='any', axis=0, inplace=True)
df['Mean'] = df.mean(axis=1)
df.head()
```

```python
world_geo = 'custom.geo.json'
df = df.reset_index()
```

```python
lifespan_map = folium.Map(tiles='Mapbox Bright')
lifespan_map.choropleth(
    geo_data=world_geo, 
    data = df, 
    columns=['country', 'Mean'], 
    key_on='feature.properties.geounit', 
    fill_color='BuPu', 
    line_opacity=0.2,
    fill_opacity = 1,
)
lifespan_map
```
